# Split the Data >>

```python
def split_data(base_folder, train_size=0.7, val_size=0.15, test_size=0.15):
    image_folder = os.path.join(base_folder, 'images')
    label_folder = os.path.join(base_folder, 'labels')
    
    for split in ['train', 'val', 'test']:
        for subfolder in ['images', 'labels']:
            os.makedirs(os.path.join(base_folder, split, subfolder), exist_ok=True)

    image_files = [f for f in os.listdir(image_folder)]
    
    #print(image_files[0])
    train_files, temp_files = train_test_split(image_files, test_size=(val_size + test_size), random_state=42)
    val_files, test_files = train_test_split(temp_files, test_size= test_size/(val_size + test_size), random_state=42)

    def move_files(file_list, split): 
        for file_name in file_list:
            src_image = os.path.join(image_folder, file_name)
            dst_image = os.path.join(base_folder, split, 'images', file_name)
            shutil.copy(src_image, dst_image)
            
            label_file_name = file_name.replace('.jpg', '.txt')
            src_label = os.path.join(label_folder, label_file_name)
            dst_label = os.path.join(base_folder, split, 'labels', label_file_name)
            if os.path.exists(src_label): 
                shutil.copy(src_label, dst_label)

    move_files(train_files, 'train')
    move_files(val_files, 'val')
    move_files(test_files, 'test')

    print("Data splitting and moving completed.")
```

```python
split_data('yolo_data')
```
